# Import Libraries

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn import naive_bayes
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from sklearn.metrics import accuracy_score, confusion_matrix , f1_score
```

# Load the Dataset

```python
data = pd.read_csv("../Kaggle Competitions/Churn Detection in telecom/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

# Data Processing

```python
pd.set_option('display.max_columns', 21)
data.head()
```

Converting TotalCharges to numeric values

```python
data.TotalCharges = pd.to_numeric(data.TotalCharges, errors = 'coerce')
data.TotalCharges.dtype
```

Check performance for any duplicates in the dataset

```python
data.duplicated().sum()
```

Dropping customerID column values

```python
data1 = data.drop(columns = 'customerID')
data1.head()
```

```python
lb = LabelEncoder() 
data1['Churn'] = lb.fit_transform(data1.Churn)
data1.head()
```

Replacing words to a 'uniform' name

```python
data1=data1.replace({'No phone service' : 'No', 'No internet service': 'No'})
data1.head(20)
```

Calculate & plot correlation values of the different variables using heatmap

```python
correl = data1.corr()
sns.heatmap(correl, annot=True, cmap="YlGnBu", linewidths=5)
```

Checking for the correlation value between two  sample variables(Churn & tenure) from the data-set

```python
print('Correlation: %0.4f' % data1.Churn.corr(data1.tenure))

```

Converting word/strings values in the dataset into dummy numerical values 

```python
data1_dummy = pd.get_dummies(data1)
data1_dummy.head(15)
data1_dummy.dropna(inplace=True)
data1_dummy.isnull().sum()
```

Correlation computation of variable Churn and the rest of the features

```python
churn_corr = data1_dummy.corr().Churn
churn_corr.sort_values(ascending = False)
```

```python
plt.figure(figsize=(16,20))
churn_corr.sort_values(ascending = False).plot(kind='bar')
```

# Data Partitioning into training and testing

```python
X = data1_dummy.drop(columns = 'Churn')
y = data1_dummy.Churn
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=42)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Model creation using Support Vector Machine(SVM)

Fitting SVM model to the training set

```python
mart_svm = SVC(kernel = 'linear', C=1)

scores = cross_val_score(mart_svm, X_train, y_train, cv=10)
print('Accuracy: %0.4f (+/- %0.2f)' %(scores.mean()*100, (scores.std()*100)*2))
```

# Model prediction

```python
mart_pred = SVC(kernel = 'linear', C=1).fit(X_train, y_train).predict(X_test)
```

# Model Validation

```python
cm = confusion_matrix(y_test, mart_pred)
print('confusion accuracy: %0.4f' %((cm.trace()/cm.sum())*100))
```

```python
print('accuracy score: %0.4f' %(accuracy_score(y_test, mart_pred)*100))
```

```python
print('f1 score: %0.4f' %(f1_score(y_test, mart_pred)*100))
```

# Model creation using Logistic Regression

Fitting Logistic Regression to the training set

```python
regressor = LogisticRegression(solver= 'lbfgs')
scores = cross_val_score(regressor, X_train, y_train, cv= 10)
print('Accuracy: %0.4f (+/- %0.2f)' %(scores.mean()*100, (scores.std()*100)*2))
```

```python
y_pred = LogisticRegression(solver= 'lbfgs').fit(X_train, y_train).predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print('confusion accuracy: %0.4f' %((cm.trace()/cm.sum())*100))
```

```python
print('accuracy score: %0.4f' %(accuracy_score(y_test, y_pred)*100))
```

```python
print('f1 score: %0.4f' %(f1_score(y_test, y_pred)*100))
```
