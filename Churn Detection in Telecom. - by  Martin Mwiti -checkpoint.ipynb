# Data Partitioning into training and testing

```python
X = data1_dummy.drop(columns = 'Churn')
y = data1_dummy.Churn
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=42)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Model creation using Support Vector Machine(SVM)

# Model creation using KNearest Neighbors(KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#search for an optimal value of K for KNN
k_score = []
for k in range(1,31):
    knn = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(knn, X, y, cv=10)
    k_score.append(scores.mean())
print (k_score)

#plot the value of K for KNN versus the Cross-Validation accuracy
k_range = range(1,31)
plt.plot(k_range , k_score)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validation Accuracy')
```

```python
#fitting KNN using the optimal value of k
knn = KNeighborsClassifier(n_neighbors = 23)
scores = cross_val_score(knn, X, y, cv=10)
print('cross validation score: %0.2f%% (+/-%0.2f%%)' %(scores.mean()*100 , (scores.std()*100)*2))
```

```python
#fitting KNN using the optimal value of k
knn = KNeighborsClassifier(n_neighbors = 23).fit(X_train, y_train)
y_pred = knn.predict(X_test)

#model validation
cm = confusion_matrix(y_test, y_pred)

#Plot confusion matrix
ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax, fmt = ',', linecolor= 'lightseagreen', linewidths=3, cmap="magma")

ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['Not Churn', ' Churn']); ax.yaxis.set_ticklabels(['Not Churn', ' Churn']);

FN = cm[1][0]
print('Minimize False Negative: ', FN )
print('Accuracy score: %0.2f%%' %(accuracy_score(y_test, y_pred)*100));print('confusion accuracy: %0.2f%%' %((cm.trace()/cm.sum())*100))
```

```python
from sklearn.ensemble import BaggingClassifier

knn = KNeighborsClassifier(n_neighbors = 23)
bc = BaggingClassifier(knn, n_estimators=10, max_samples=0.6, max_features=1.0, bootstrap=True, bootstrap_features=False)
bc.fit(X_train, y_train)
print('Bagging Score: %0.2f%% ' %(bc.score(X_test, y_test)*100))
```

# Model creation using Logistic Regression

```python
lr = LogisticRegression(solver= 'lbfgs', max_iter=1000)
scores = cross_val_score(lr, X, y, cv= 10)
print('Accuracy: %0.2f%% (+/- %0.2f%%)' %(scores.mean()*100, (scores.std()*100)*2))
```

```python
lr = LogisticRegression(solver= 'lbfgs', max_iter=1000).fit(X_train, y_train)
y_pred = lr.predict(X_test)

#model validation
cm = confusion_matrix(y_test, y_pred)

#Plot confusion matrix
sns.heatmap(cm, annot=True, ax = ax, fmt = ',', linecolor= 'lightseagreen', linewidths=3, cmap="magma")

print('confusion accuracy: %0.2f%%' %((cm.trace()/cm.sum())*100));print('accuracy score: %0.2f%%' %(accuracy_score(y_test, y_pred)*100))
#print('f1 score: %0.2f%%' %(f1_score(y_test, y_pred)*100))
```
